## Importação das bibliotecas

```python
import pandas            as pd


import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import warnings
warnings.filterwarnings('ignore')
```

## AULA 01 - Aquisição e leitura de dados reais

```python
# encoding == codificacao
# skiptows == não considera as x primeiras linhas
# skipfooter == não considera as x ultimas linhas
# thousands == Separação do milhar
# decimal == separacao das dezenas

dados = pd.read_csv('A092517189_28_143_208_processamento.csv',
                   encoding = 'ISO-8859-1', skiprows = 3, skipfooter = 12, sep = ';',
                   thousands = '.',
                   decimal = ',') 
```

```python
dados
```

```python
dados.tail() # 5 ultimos resultados
```

```python
# informacoes da tabela
dados.info()
```

```python
# Os números são representados em notacão cientifica
dados.mean()
```

```python
# configuracao do pandas. Neste caso, estamos alterando o "display" de float para duas casas decimais
pd.options.display.float_format = "{:.2f}".format
```

```python
# Visualizando a média após a configuração do display
dados.mean()
```

```python
# Media somente da coluna 2008/Ago
dados['2008/Ago'].mean()
```

```python
# Primeiro plot
dados.plot(x='Unidade da Federação', y='2008/Ago')
```

```python
# plot alterando o tipo de gráfico e o tamanho
dados.plot(x='Unidade da Federação', y='2008/Ago', kind = 'bar', figsize=(8,6))
```

```python

ax = dados.plot(x='Unidade da Federação', y='2008/Ago', kind = 'bar', figsize=(8,6))

# Configuracao do eixo Y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

#Titulo do gráfico
plt.title('Valor por unidade da Federação')
```

### Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.

```python

ax = dados.sort_values(by='2008/Ago', ascending=False).plot(x='Unidade da Federação', y='2008/Ago', kind = 'bar', figsize=(12,5))

# Configuracao do eixo Y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

#Titulo do gráfico
plt.title('GASTOS HOSPITALARES PROCESSADOS EM 08/2008 POR UNIDADE DA FEDERAÇÃO',
         pad = 20, fontsize = 14, color = 'dimgrey', fontweight = 'bold')

# Configuracao do eixo X e Y
plt.xlabel('Unidade da Federação', color = 'grey', labelpad = 20, fontsize = 10)
plt.ylabel('Gastos Processados', color = 'grey', labelpad = 20, fontsize = 10)
plt.yticks(fontsize = 10, color = 'dimgrey')
plt.xticks(fontsize = 10, color = 'dimgrey')

# Remoção das bordas
plt.box(False)


# Removendo a legenda
ax.get_legend().remove()


plt.show()
```
