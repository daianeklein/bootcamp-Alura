## Aula 02 - Primeiras Visualizações de Dados

```python
# Com a funcao mean, o Pandas traz somente os meses que possuem valor (há dados nulos no df)
dados.mean()
```

```python
# Uma vez que a funcao mean só irá trazer os meses com valores, serão transformados em indices
# tolist para que os valores sejam transformados em uma lista, caso contrário, os valores ficam imutáveis
colunas_usaveis = dados.mean().index.tolist()

# inserindo a coluna 'Unidade da Federação' na variável colunas_usaveis
colunas_usaveis.insert(0, 'Unidade da Federação')
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head(3)
```

```python
# Definindo a coluna 'Unidade da Federação' como índice
usaveis = usaveis.set_index('Unidade da Federação')
usaveis.head(3)
```

```python
# Localizando um valor pelo indice
usaveis.loc['12 Acre']
```

```python
# Transpondo o DataFrame
usaveis.T.head(3)
```

```python
# O grafico abaixo ficou com o eixo Y bastante deslocado, pois na variável 'usaveis' o total está sendo considerado
usaveis.T.plot(figsize=(12,5))
```

```python
# Excluindo o 'Total' da variável 'usaveis'
usaveis = usaveis.drop('Total', axis = 1)
usaveis.head(2)
```

```python
# Plotando novamente o gráfico
usaveis.T.plot(figsize=(12,6))
```

### Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada

```python
# Para configurar o eixo Y, será criada uma variável com o maior valor
maior_valor = usaveis.max(axis=1).max()

# Plotando gráfico
ax = usaveis.T.plot(figsize=(14,6))

# Configuracao do eixo Y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))


#Titulo do gráfico
plt.title('GASTOS HOSPITALARES ATENDIDOS POR UNIDADE DA FEDERAÇÃO NO PERÍODO DE AGO/07 à JUL/20',
         pad = 20, fontsize = 14, color = 'dimgrey', fontweight = 'bold')

# Configuracao do eixo X e Y
plt.xlabel('MM/YY', color = 'grey', labelpad = 20, fontsize = 10)
plt.ylabel('Gastos Atendidos', color = 'grey', labelpad = 20, fontsize = 10)
plt.yticks(fontsize = 10, color = 'dimgrey')
plt.xticks(fontsize = 10, color = 'dimgrey')
plt.xticks(rotation=90)

# Remoção das bordas
plt.box(False)

# "Movendo" a legenda
plt.legend(loc=(1.03, -0.1), borderaxespad=0)

# Alterando o eixo Y (comeca do 0, limite será o maior valor do dataset (variavel maior_valor), e 
# o eixo "pula" de 100.000.000 em 100.000.000)
y_ticks = np.arange(0, maior_valor, 100000000)
plt.yticks(y_ticks, color = 'grey', fontsize = 8)

plt.show()
```

### Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
#Selecionando 5 estados aleatorios
usaveis_sample = usaveis.sample(5, random_state = 55)

# Para configurar o eixo Y, será criada uma variável com o maior e menor valor
maior_valor2 = usaveis_sample.max(axis=1).max()
menor_valor2 = usaveis_sample.min(axis=1).min()

# Plotando gráfico (funcao sample para plotar 5 estados)
ax = usaveis_sample.T.plot(figsize=(14,6))

# Configuracao do eixo Y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))


#Titulo do gráfico
plt.title('GASTOS HOSPITALARES ATENDIDOS POR UNIDADE DA FEDERAÇÃO NO PERÍODO DE AGO/07 à JUL/20',
         pad = 20, fontsize = 14, color = 'dimgrey', fontweight = 'bold')

# Configuracao do eixo X e Y
plt.xlabel('MM/YY', color = 'grey', labelpad = 20, fontsize = 10)
plt.ylabel('Gastos Atendidos', color = 'grey', labelpad = 20, fontsize = 10)
plt.yticks(fontsize = 10, color = 'dimgrey')
plt.xticks(fontsize = 10, color = 'dimgrey')
plt.xticks(rotation=90)

# Remoção das bordas
plt.box(False)

# "Movendo" a legenda
plt.legend(loc=(1.03, -0.1), borderaxespad=0)

# Alterando o eixo Y (comeca do 0, limite será o maior valor do dataset (variavel maior_valor), e 
# o a escala do eixo Y será de 10000000 em 10000000
y_ticks = np.arange(0, maior_valor2, 10000000)
plt.yticks(y_ticks, color = 'grey', fontsize = 8)

plt.show()
```

```python
#Somando cada um dos estados, ou seja, somando as linhas e criando a coluna total
usaveis['Total'] = usaveis.sum(axis=1)
usaveis.head(2)
```

```python
# Fazendo a ordenacao por total e excluindo a coluna
ordenado_por_total = usaveis.sort_values(by='Total', ascending=False)
ordenado_por_total = ordenado_por_total.drop('Total', axis = 1)
ordenado_por_total.head(2)
```

```python
# plot do grafico ordenado por total
ordenado_por_total.head(5).T.plot(figsize=(10,6))
```

```python
# removendo a coluna com gastos muito elevados
colunas_interessadas = ordenado_por_total.columns[6:]
ordenado_por_total = ordenado_por_total[colunas_interessadas]
```

```python
# Plot do gráfico
ordenado_por_total.head(5).T.plot(figsize=(10,6))
```

### Desafio 01: Escolher uma palete de cores mais adequada do matplotlib.

```python
# será utilizado a varíavel 'ordenado_por_total'
ax = ordenado_por_total.sample(5, random_state = 35).T.plot(figsize=(13, 5), colormap = 'plasma')


# Configuracao do eixo Y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
 

#Titulo do gráfico
plt.title('GASTOS HOSPITALARES ATENDIDOS POR UNIDADE DA FEDERAÇÃO NO PERÍODO A PARTIR DE 02/2008',
         pad = 20, fontsize = 14, color = 'dimgrey', fontweight = 'bold')

# Configuracao do eixo X e Y
plt.xlabel('MM/YY', color = 'grey', labelpad = 20, fontsize = 10)
plt.ylabel('Gastos Atendidos', color = 'grey', labelpad = 20, fontsize = 10)
plt.yticks(fontsize = 10, color = 'dimgrey')
plt.xticks(fontsize = 10, color = 'dimgrey')
plt.xticks(rotation=90)

# Remoção das bordas
plt.box(False)

# "Movendo" a legenda
plt.legend(loc=(1.13, -0.2), borderaxespad=0)

# Configurando o eixo Y
# Obtendo o maior valor
maior_valor = ordenado_por_total.sample(5, random_state = 35).max().max()

y_ticks = np.arange(0, maior_valor, 15000000)
plt.yticks(y_ticks, color = 'grey', fontsize = 8)

plt.show()
```

### Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Suldeste e Centro-Oeste) de cada estado.

```python
# extraindo a primeira "letra" da string 
#regioes = ordenado_por_total.index.str[0]

# Dicionario com as regiões
rg = {'1':'norte', '2':'nordeste', '3':'sudeste', '4':'sul', '5':'centro-oeste'}

# Funcao map para criar as colunas
# index.str[0] fará um "slice" bucando a primeira letra da coluna 'Unidade da Federacao'
# map irá fazer a busca dos valores
ordenado_por_total['regiao'] = ordenado_por_total.index.str[0].map(rg)

```

```python
# Conferindo
ordenado_por_total['regiao'].sample(3)
```
